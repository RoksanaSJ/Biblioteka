```python
#Replacing null values ​​with 'Unknown' values.
airbnb['NAME'] = airbnb['NAME'].fillna('Unknown')
airbnb['host id'] = airbnb['host id'].fillna('Unknown')
airbnb['host_identity_verified'] = airbnb['host_identity_verified'].fillna('Unknown')
airbnb['host name'] = airbnb['host name'].fillna('Unknown')
airbnb['neighbourhood group'] = airbnb['neighbourhood group'].fillna('Unknown')
airbnb['neighbourhood'] = airbnb['neighbourhood'].fillna('Unknown')
airbnb['lat'] = airbnb['lat'].fillna('Unknown')
airbnb['country'] = airbnb['country'].fillna('Unknown')
airbnb['country code'] = airbnb['country code'].fillna('Unknown')
airbnb['instant_bookable'] = airbnb['instant_bookable'].fillna('Unknown')
airbnb['cancellation_policy'] = airbnb['cancellation_policy'].fillna('Unknown')
airbnb['room type'] = airbnb['room type'].fillna('Unknown')
airbnb['Construction year'] = airbnb['Construction year'].fillna('Unknown')
airbnb['price'] = airbnb['price'].fillna('Unknown')
airbnb['service fee'] = airbnb['service fee'].fillna('Unknown')
airbnb['minimum nights'] = airbnb['minimum nights'].fillna('Unknown')
airbnb['number of reviews'] = airbnb['number of reviews'].fillna('Unknown')
airbnb['reviews per month'] = airbnb['reviews per month'].fillna('Unknown')
airbnb['review rate number'] = airbnb['review rate number'].fillna('Unknown')
airbnb['calculated host listings count'] = airbnb['calculated host listings count'].fillna('Unknown')
airbnb['availability 365'] = airbnb['availability 365'].fillna('Unknown')
airbnb['house_rules'] = airbnb['house_rules'].fillna('Unknown')
airbnb['license'] = airbnb['license'].fillna('Unknown')
airbnb
```

```python
#I added a column that categorizes the 'price' column.
def categorize_price(price):
    try:
        if isinstance(price, str):
            price = price.replace('$', '').replace(',', '')
        p = float(price)
    except (ValueError, TypeError):
        return "Unknown"
    if p >= 800:
        return "Expensive"
    elif p > 400:
        return "Medium"
    else:
        return "Cheap"

airbnb['Is_Expensive'] = airbnb['price'].apply(categorize_price)
airbnb
```

```python
#I created a new column that categorizes data by the 'number of reviews' column.
def categorize_number_of_reviews(number_of_reviews):
    try:
        if isinstance(number_of_reviews, str):
            number_of_reviews = number_of_reviews.replace('$', '').replace(',', '')
        p = float(number_of_reviews)
    except (ValueError, TypeError):
        return "Unknown"
    if p > 512:
        return "Very popular"
    elif p >= 512:
        return "Popular"
    elif p >= 256:
        return "Medium interest"
    else:
        return "Low interest"

airbnb['Is popular'] = airbnb['number of reviews'].apply(categorize_number_of_reviews)
airbnb
```

```python
#I created a new column that categorizes data by the 'minimum nights' column.
def categorize_min_nights (min_nights):
    try:
        if isinstance(min_nights, str):
          min_nights = min_nights.replace('$', '').replace(',', '')
        p = float(min_nights)
    except (ValueError, TypeError):
        return "Unknown"
    if p < 1:
      return "Error"
    elif p > 1 and p <= 30:
      return "Short term rental"
    elif p > 30:
      return "Long term rental"
    else:
      return "Unknown"

airbnb['Is Long Term'] = airbnb['minimum nights'].apply(categorize_min_nights)
airbnb
```

```python
#I created a new column that categorizes data by the 'review rate number' column.

def categorize_review_rate_number (review_rate_number):
    try:
        if isinstance(review_rate_number, str):
          review_rate_number = review_rate_number.replace('$', '').replace(',', '')
        p = float(review_rate_number)
    except (ValueError, TypeError):
        return "Unknown"
    if p == 5:
      return "Very Good"
    elif p >= 4:
      return "Good"
    elif p >= 3:
      return "Average"
    elif p >= 1:
      return "Very Bad"
    else:
      return "Unknown"

airbnb['Review Rate Category'] = airbnb['review rate number'].apply(categorize_review_rate_number)
airbnb
```

```python
#I created a new column that categorizes data by the 'Construction year' column.

def categorize_construction_year (construction_year):
    try:
        if isinstance(construction_year, str):
          construction_year = construction_year.replace('$', '').replace(',', '')
        y = float(construction_year)
    except (ValueError, TypeError):
        return "Unknown"
    if y >= 2019:
        return "Brand New (2019-2022)"
    elif y >= 2011:
        return "Modern (2011-2018)"
    elif y >= 2003:
        return "Early 2000s (2003-2010)"
    else:
        return "Older"

airbnb['Construction Year Category'] = airbnb['Construction year'].apply(categorize_construction_year)
airbnb
```

```python
#I cleaned up the data with the incorrect name in the 'neighborhood' group' column.
airbnb['neighbourhood group'] = airbnb['neighbourhood group'].replace("manhatan", "Manhattan")
airbnb['neighbourhood group'] = airbnb['neighbourhood group'].replace("brookln", "Brooklyn")
airbnb
```

```python
#Changing the data type in numeric columns

airbnb['lat'] = pd.to_numeric(airbnb['lat'], errors='coerce')
airbnb['long'] = pd.to_numeric(airbnb['long'], errors='coerce')
airbnb['Construction year'] = pd.to_numeric(airbnb['Construction year'], errors='coerce')
airbnb['price'] = pd.to_numeric(airbnb['price'].replace(r'[\$,\s]', '', regex=True), errors='coerce')
airbnb['service fee'] = pd.to_numeric(airbnb['service fee'].replace(r'[\$,\s]', '', regex=True), errors='coerce')
airbnb['minimum nights'] = pd.to_numeric(airbnb['minimum nights'], errors='coerce')
airbnb['number of reviews'] = pd.to_numeric(airbnb['number of reviews'], errors='coerce')
airbnb['reviews per month'] = pd.to_numeric(airbnb['reviews per month'], errors='coerce')
airbnb['review rate number'] = pd.to_numeric(airbnb['review rate number'], errors='coerce')
airbnb['calculated host listings count'] = pd.to_numeric(airbnb['calculated host listings count'], errors='coerce')
airbnb['availability 365'] = pd.to_numeric(airbnb['availability 365'], errors='coerce')
```

```python
#Boxplot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

sns.boxplot(data=airbnb, x='neighbourhood group', y='price', palette='Spectral', fliersize=2)

plt.title('Price distribution in districts', fontsize=15)
plt.xlabel('Neighbourhood', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.xticks(rotation=45)

plt.show()
```

```python
#Average price by district and room type
plt.figure(figsize=(12, 7))
sns.heatmap(price_map, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Average price by district and room type')
plt.show()
print('\n')

#Impact of rating on price
plt.figure(figsize=(10, 6))
sns.barplot(data=airbnb, x='review rate number', y='price', palette='magma')
plt.title('Impact of rating on price')
plt.show()
print('\n')

#Popularity analysis (reviews per month) vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=airbnb, x='price', y='reviews per month', alpha=0.1)
plt.title('Price vs. Monthly Reviews')
plt.show()
print('\n')

#What correlates most with price?
plt.figure(figsize=(10, 8))
numeric_cols = airbnb.select_dtypes(include=['number']).corr()
sns.heatmap(numeric_cols[['price']].sort_values(by='price', ascending=False), annot=True, cmap='coolwarm')
plt.title('What correlates most with price?')
plt.show()
print('\n')

#Room Type vs. Price
plt.figure(figsize=(10,6))
sns.barplot(data=airbnb, x='room type', y='price', palette='viridis')
plt.title('Room Type vs. Price')
plt.show()
print('\n')

#Price distribution in neighbourhoods
plt.figure(figsize=(12,6))
sns.barplot(data=airbnb, x='neighbourhood group', y='price')
plt.title('Price distribution in neighbourhoods')
plt.show()
```
